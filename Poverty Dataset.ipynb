```python
#dependencies
import json
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from census import Census
from config import api_key
```

```python
c_2011 = Census(api_key, year=2011)
census_data = c_2011.acs5.get(("B25064_001E", "B01003_001E", "B19013_001E", "B17001_002E"), {'for': 'zip code tabulation area:*'})

census_2011 = pd.DataFrame(census_data)
census_2011

# Column Reordering
census_2011 = census_2011.rename(columns={"zip code tabulation area": "Zipcode",
                                          "B25064_001E": "2011 Median Rent+Utilities",
                                      "B01003_001E": "2011 Total Population",
                                      "B19013_001E": "2011 Household Income",
                                      "B17001_002E": "2011 Poverty Count",
                                      "B01002_001E": "2011 Median Age",
                                      })
#add in Poverty rate
census_2011['2011 Poverty Rate'] = census_2011['2011 Poverty Count'] / census_2011['2011 Total Population'] *100


# Final DataFrame
census_2011 = census_2011[["Zipcode", "2011 Total Population", "2011 Household Income",
                       "2011 Median Rent+Utilities", "2011 Poverty Count", "2011 Poverty Rate"]]
census_2011
```
