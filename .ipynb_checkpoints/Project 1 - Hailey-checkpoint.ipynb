```python
import json
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from census import Census
from config import api_key
```

```python
#trying to find data from another year - There is not Census info for each year

c_2011 = Census(api_key, year=2011)
census_data = c_2011.acs5.get(("NAME", "B25064_001E", "B01003_001E", "B19013_001E", "B23025_005E", "B01002_001E", "B17001_002E"), {'for': 'zip code tabulation area:*'})

census_2011 = pd.DataFrame(census_data)

# Column Reordering
census_2011 = census_2011.rename(columns={"B25064_001E": "Median Rent+Utilities 2011",
                                      "B01003_001E": "Total Population 2011",
                                      "B19013_001E": "Household Income 2011",
                                      "B23025_005E": "# Unemployed 2011",
                                      "B17001_002E": "Poverty Count 2011",
                                      "B01002_001E": "Median Age 2011",
                                      "NAME": "Name", "zip code tabulation area": "Zipcode"})
#add in unemployment rate
census_2011['Unemployment Rate 2011'] = census_2011['# Unemployed 2011'] / census_2011['Total Population 2011'] *100

#add in Poverty rate
census_2011['Poverty Rate 2011'] = census_2011['Poverty Count 2011'] / census_2011['Total Population 2011'] *100

# Final DataFrame
census_2011 = census_2011[["Zipcode", "Total Population 2011", "Household Income 2011",
                       "Median Rent+Utilities 2011", "# Unemployed 2011", "Unemployment Rate 2011",'Poverty Rate 2011']]

#drop negative values
census_2011 = census_2011.loc[census_2011["Median Rent+Utilities 2011"] > 0, :]

# Visualize
census_2011.head(10)
```

```python
#pulling data from 2015
c_2015 = Census(api_key, year=2015)
census_data = c_2015.acs5.get(("NAME", "B25064_001E", "B01003_001E", "B19013_001E", "B23025_005E", "B01002_001E", "B17001_002E"), {'for': 'zip code tabulation area:*'})

census_2015 = pd.DataFrame(census_data)

# Column Reordering
census_2015 = census_2015.rename(columns={"B25064_001E": "Median Rent+Utilities 2015",
                                      "B01003_001E": "Total Population 2015",
                                      "B19013_001E": "Household Income 2015",
                                      "B23025_005E": "# Unemployed 2015",
                                      "B17001_002E": "Poverty Count 2015",
                                      "B01002_001E": "Median Age 2015",
                                      "NAME": "Name", "zip code tabulation area": "Zipcode"})
#add in unemployment rate
census_2015['Unemployment Rate 2015'] = census_2015['# Unemployed 2015'] / census_2015['Total Population 2015'] *100

#add in Poverty rate
census_2015['Poverty Rate 2015'] = census_2015['Poverty Count 2015'] / census_2015['Total Population 2015'] *100

# Final DataFrame
census_2015 = census_2015[["Zipcode", "Total Population 2015", "Household Income 2015",
                       "Median Rent+Utilities 2015", "# Unemployed 2015", "Unemployment Rate 2015", 'Poverty Rate 2015']]


#drop negative values
census_2015 = census_2015.loc[census_2015["Median Rent+Utilities 2015"] > 0, :]


# Visualize
print(len(census_2015))
census_2015.head(10)
```

```python
## merge all years of data
merge_11_12= pd.merge(census_2011, census_2012, on="Zipcode", how="left")
merge_with_13 = pd.merge(merge_11_12, census_2013, on='Zipcode', how='left')
merge_with_14 = pd.merge(merge_with_13, census_2014, on='Zipcode', how='left')
all_data = pd.merge(merge_with_14, census_2015, on='Zipcode', how='left')
all_data=all_data.dropna(how='any')
all_data
```

```python
#rent costs dataframe
rent_costs_year = all_data[['Zipcode', 'Median Rent+Utilities 2011', 'Median Rent+Utilities 2012', 'Median Rent+Utilities 2013', 'Median Rent+Utilities 2014', 'Median Rent+Utilities 2015']]
rent_cost_compare = pd.DataFrame(rent_costs_year)
rent_cost_compare.set_index(['Zipcode'])
rent_cost_compare
```

```python
# unemployment rate
unemp_pct_years = all_data[['Zipcode', 'Unemployment Rate 2011', 'Unemployment Rate 2012', 'Unemployment Rate 2013', 'Unemployment Rate 2014', 'Unemployment Rate 2015']]
unempployment_rate = pd.DataFrame(unemp_pct_years)
unempployment_rate.set_index(['Zipcode'])
unempployment_rate
```

```python
# income
household_income_years = all_data[['Zipcode', 'Household Income 2011', 'Household Income 2012', 
                            'Household Income 2013', 'Household Income 2014', 'Household Income 2015']]
household_income = pd.DataFrame(household_income_years)
household_income.set_index(['Zipcode'])
household_income
```

```python
#povery rate table
poverty_data = all_data[["Zipcode", "Poverty Rate 2011", "Poverty Rate 2012", "Poverty Rate 2013", "Poverty Rate 2014", "Poverty Rate 2015"]]
poverty_data
```

```python
#get lats and longs
cities_info_df = pd.read_csv("zips_lats_longs.csv", sep=";", dtype="object", encoding="utf-8")
cities_info_df = cities_info_df.rename(columns={"Zip": "Zipcode"})
cities_info_df = cities_info_df.loc[:, ["Zipcode", "City", "State", "Latitude", "Longitude"]]
# Visualize
cities_info_df.head()
```

```python
#merge zips with lats and longs
ziplatlong = pd.merge(all_data, cities_info_df, on='Zipcode', how='left')
ziplatlong
```

```python
ziplatlong.columns
```

```python
# define poverty levels each year
poverty_years = 2011, 2012, 2013, 2014, 2015
poverty_levels = 10890, 11170, 11490, 11670, 11770
```

```python
avg_annual_income_raw = ziplatlong[['Household Income 2011','Household Income 2012','Household Income 2013',
                                'Household Income 2014','Household Income 2015']].mean()

avg_annual_income = pd.DataFrame(avg_annual_income_raw).T

avg_annual_income.rename(columns = {'Household Income 2011':'2011','Household Income 2012':'2012','Household Income 2013':'2013',
                                'Household Income 2014':'2014','Household Income 2015':'2015'}, inplace = True)

avg_income_rent = avg_annual_income.T

avg_income_rent.rename(columns = {0:'Household_Income'}, inplace = True)
                                  
avg_income_rent
```

```python
ziplatlong['Annual Rent+Utilities 2011'] = ziplatlong['Median Rent+Utilities 2011'] * 12
ziplatlong['Annual Rent+Utilities 2012'] = ziplatlong['Median Rent+Utilities 2012'] * 12
ziplatlong['Annual Rent+Utilities 2013'] = ziplatlong['Median Rent+Utilities 2013'] * 12
ziplatlong['Annual Rent+Utilities 2014'] = ziplatlong['Median Rent+Utilities 2014'] * 12
ziplatlong['Annual Rent+Utilities 2015'] = ziplatlong['Median Rent+Utilities 2015'] * 12

avg_rent_utilities_raw = ziplatlong[['Annual Rent+Utilities 2011','Annual Rent+Utilities 2012','Annual Rent+Utilities 2013',
                                'Annual Rent+Utilities 2014','Annual Rent+Utilities 2015']].mean()

avg_rent_utilities = pd.DataFrame(avg_rent_utilities_raw).T

avg_rent_utilities.rename(columns = {'Annual Rent+Utilities 2011':'2011','Annual Rent+Utilities 2012':'2012','Annual Rent+Utilities 2013':'2013',
                                'Annual Rent+Utilities 2014':'2014','Annual Rent+Utilities 2015':'2015'}, inplace = True)

avg_rent_utilities = avg_rent_utilities.T

avg_rent_utilities.rename(columns = {0:'Rent'}, inplace = True)

avg_rent_utilities
```

```python
avg_unemployment_rate_raw = ziplatlong[['Unemployment Rate 2011','Unemployment Rate 2012','Unemployment Rate 2013',
                                'Unemployment Rate 2014','Unemployment Rate 2015']].mean()

avg_unemployment_rate = pd.DataFrame(avg_unemployment_rate_raw).T

avg_unemployment_rate.rename(columns = {'Unemployment Rate 2011':'2011','Unemployment Rate 2012':'2012','Unemployment Rate 2013':'2013',
                                'Unemployment Rate 2014':'2014','Unemployment Rate 2015':'2015'}, inplace = True)

avg_unemployment_rate = avg_unemployment_rate.T

avg_unemployment_rate.rename(columns = {0:'Unemployment_Rate'}, inplace = True)

avg_unemployment_rate
```

```python
avg_poverty_rate_raw = ziplatlong[['Poverty Rate 2011','Poverty Rate 2012','Poverty Rate 2013',
                                'Poverty Rate 2014','Poverty Rate 2015']].mean()

avg_poverty_rate = pd.DataFrame(avg_poverty_rate_raw).T

avg_poverty_rate.rename(columns = {'Poverty Rate 2011':'2011','Poverty Rate 2012':'2012','Poverty Rate 2013':'2013',
                                'Poverty Rate 2014':'2014','Poverty Rate 2015':'2015'}, inplace = True)

avg_poverty_rate = avg_poverty_rate.T

avg_poverty_rate.rename(columns = {0:'Poverty_Rate'}, inplace = True)

avg_poverty_rate
```

```python
ziplatlong['Affordability 2011'] = ziplatlong['Annual Rent+Utilities 2011'] / ziplatlong['Household Income 2011']
ziplatlong['Affordability 2012'] = ziplatlong['Annual Rent+Utilities 2012'] / ziplatlong['Household Income 2012']
ziplatlong['Affordability 2013'] = ziplatlong['Annual Rent+Utilities 2013'] / ziplatlong['Household Income 2013']
ziplatlong['Affordability 2014'] = ziplatlong['Annual Rent+Utilities 2014'] / ziplatlong['Household Income 2014']
ziplatlong['Affordability 2015'] = ziplatlong['Annual Rent+Utilities 2015'] / ziplatlong['Household Income 2015']
```

```python
avg_affordability_raw = ziplatlong[['Affordability 2011','Affordability 2012','Affordability 2013',
                                'Affordability 2014','Affordability 2015']].mean()

avg_affordability = pd.DataFrame(avg_affordability_raw).T

avg_affordability.rename(columns = {'Affordability 2011':'2011','Affordability 2012':'2012','Affordability 2013':'2013',
                                'Affordability 2014':'2014','Affordability 2015':'2015'}, inplace = True)

avg_affordability = avg_affordability.T

avg_affordability.rename(columns = {0:'Affordability'}, inplace = True)

avg_affordability
```

```python
avg_income_rent['Rent'] = avg_rent_utilities['Rent']

avg_income_rent['Unemployment_Rate'] = avg_unemployment_rate['Unemployment_Rate']

avg_income_rent['Poverty_Rate'] = avg_poverty_rate['Poverty_Rate']

avg_income_rent['Affordability'] = avg_affordability['Affordability']

avg_income_rent = avg_income_rent.reset_index()

avg_income_rent.rename(columns = {'index':'Year'}, inplace = True)

avg_income_rent
```

```python
fig,ax=plt.subplots()
ax.plot(avg_income_rent.Year, avg_income_rent.Rent, marker="o", color="b")
ax.set_xlabel("Year")
ax.set_ylabel("Rent ($ Annual)", color='b')
ax.set_ylim(bottom=8000,top=11000)
ax2=ax.twinx()
ax2.plot(avg_income_rent.Year, avg_income_rent.Unemployment_Rate, marker="o", color="r")
ax2.set_ylabel("Unemployment Rate (%)", color='r')
ax2.set_ylim(bottom=2.5,top=7)
plt.title("Annual Rent vs Unemployment Rate")
for tl in ax.get_yticklabels():
    tl.set_color('b')
for tl in ax2.get_yticklabels():
    tl.set_color('r')


plt.show()
```
